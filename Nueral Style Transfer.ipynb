```python
def load_image(img_path):
    
    img = tf.io.read_file(img_path)
    img = tf.image.decode_image(img, channels=3)
    img = tf.image.convert_image_dtype(img, tf.float32)
    img = img[tf.newaxis, :]
    
    return img
```

```python
content_image = load_image('afghan women.jpg')
style_image = load_image('iran blood 2.jpg')
```

```python
plt.imshow(np.squeeze(style_image))
plt.show()
```

```python
stylized_img = model(tf.constant(content_image), tf.constant(style_image))
```

```python
plt.imshow(np.squeeze(stylized_img))
plt.show()
```
